```python
from diffusers import StableDiffusion3Pipeline

pipe = StableDiffusion3Pipeline.from_pretrained("stabilityai/stable-diffusion-3.5-medium", torch_dtype=torch.bfloat16)
pipe = pipe.to("cuda")
```

```python
image = pipe(
    "an androgynous noisey face made of multiple eyes, blackhole, floating in space, 90s noise music aesthetic. oil painting, in style of marcel duchamp surrealism, 21st century schizoid man",
    num_inference_steps=40,
    guidance_scale=4.5,
).images[0]
image.save("kamprompt.png")
```
